```python
import matplotlib.pyplot as plt

midpoint = len(df1) // 2

df1_first_half = df1.iloc[:midpoint]
plt.figure(figsize=(8, 8))
df1_first_half['STATE/UT'].value_counts().plot(kind='pie', autopct="%.2f", fontsize=9)
plt.title("PERCENTAGE STATES/UTs AMONG DIFFERENT CLASSES OF AVERAGE LITERACY RATES (First Half)")
plt.ylabel('')  
plt.show()
```

```python
import matplotlib.pyplot as plt

midpoint = len(df1) // 2

df1_first_half = df1.iloc[midpoint:]

plt.figure(figsize=(6, 6))

df1_first_half['STATE/UT'].value_counts().plot(kind='pie', autopct="%.2f", fontsize=9)
plt.title("PERCENTAGE STATES/UTs AMONG DIFFERENT CLASSES OF AVERAGE LITERACY RATES (First Half)")
plt.ylabel('') 
plt.show()
```

```python
plt.figure(figsize=(8,8),facecolor='Yellow')
sns.barplot(x=df1['DISTRICT'],y=df1['No. of years'])
plt.ylabel("No. of years")
plt.title("District vs years")
plt.xticks(rotation=90)

plt.grid()
plt.show()
```

From the above graph we can see that, Banglore, Patna, Tripura has murder cases for 3 or more than 3 years.
